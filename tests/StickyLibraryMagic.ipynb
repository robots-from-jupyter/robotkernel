```robotframework
%%python module LibraryWithGlobalState

class LibraryWithGlobalState:
    ROBOT_LIBRARY_SCOPE = 'GLOBAL'
        
    def __init__(self):
        self.state = 0
        
    def get_increased_global_state(self):
        self.state += 1
        return self.state
```
